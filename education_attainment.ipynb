### Profile Report Conclusions

Findings from the profile report generated by ydata_profiling:

* There are 38 columns(variables) and 626 rows (observations) in the dataset.
  
* Missing values:
  * There are 18 missing values in total which is 0.1% of the number of values in the dataset.
  * Only 1 value is missing on the columns where missing values are detected

* Duplicates:
     * There is only 1 duplicated row with the country value being Panama in the year 1995  

* Correlations
    * All columns have a positive correlation.
    * The age groups of males between 30-39 have the least coefficient ratio of (0.095) in relation to the year meaning that the grades of males in this age group tend to minimally increase in comparison to other age groups every year.
